```python
import numpy as np
import pandas as pd
heroes = pd.read_csv("heroes_information.csv")
```

```python
###instalar desde anaconda navigator######
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
query1 = " SELECT * FROM heroes limit 10"
```

```python
pysqldf(query1)
```

```python
query2 = "select name from heroes limit 10"
```

```python
pysqldf(query2)
```

```python
query3 = "select distinct Publisher from heroes"
pysqldf(query3)
```

```python
query4 = "select count(Height) from heroes where Height > 200"
pysqldf(query4)
```

```python
query5 = "select * from heroes where Race = 'Human' and Height > 200"
pysqldf(query5)
```

```python
query6 = "select * from heroes where Weight between 100 and 200"
pysqldf(query6)
```

```python
query7 = "select * from heroes where [Eye color] in ('red','blue')"
pysqldf(query7)
```

```python
query9 = "select count(*) from heroes where (Race in ('Human','Mutant') and [Hair color] = 'Green') or (Race = 'Vampire' and [Hair color] = 'Black') "
pysqldf(query9)
```

```python
query10 = "select name from heroes order by Race DESC limit 1"
pysqldf(query10)
```

```python
query11 = "SELECT count(*) from heroes group by Gender"
pysqldf(query11)
```

```python
query12 = "SELECT count(id) from heroes group by Publisher having count(id)>15 "
pysqldf(query12)
```
